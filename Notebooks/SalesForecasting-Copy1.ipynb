```python
import pandas as pd
```

```python
df = pd.read_csv("../data/superstore.csv", encoding="latin-1")
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df['Order Date'].head() # converted 'Order Date' column in datetime
```

```python
df[df['Sales'] > 0].head() # accessed only values that are greater than 0
```

```python
monthly_sales = (df.groupby(pd.Grouper(key='Order Date', freq='ME'))['Sales'].sum().reset_index())
# It calculates total sales for each month using the Order Date.
# freq='ME' → Month End frequency (groups all dates into months)
# pd.Grouper(key='Order Date', freq='ME') # Tells pandas to group data by month
```

```python
monthly_sales.rename(columns={'Order Date': 'ds','Sales': 'y'}, inplace=True)
```

```python
monthly_sales.head()
```

```python
Q1 = monthly_sales['y'].quantile(0.25)
Q3 = monthly_sales['y'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
monthly_sales['y'] = monthly_sales['y'].clip(lower=lower_bound, upper=upper_bound)
```

```python
import numpy as np
```

```python
monthly_sales['y_log'] = np.log1p(monthly_sales['y'])
```

```python
monthly_sales.head()
```

```python
prophet_df = monthly_sales[['ds', 'y_log']].rename(columns={'y_log': 'y'})
```

```python
prophet_df.head() # final table for fitting model
```

```python
# Now train the forecasting model (Prophet)
from prophet import Prophet

model = Prophet(yearly_seasonality=True, weekly_seasonality=False, daily_seasonality=False, changepoint_prior_scale=0.1)
model.add_seasonality(name='quarterly', period=91.25, fourier_order=8)
model.fit(prophet_df)
```

```python
future_predictions = model.make_future_dataframe(periods=6, freq='ME')
Predicted_Sales = model.predict(future_predictions)
```

```python
Predicted_Sales.head()
```

```python
forecast_output = Predicted_Sales[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].merge(monthly_sales, on='ds', how='left')
# It combines Prophet’s predicted values with the actual monthly sales into one table.
```

```python
forecast_output.rename(columns={'ds': 'Date', 'y': 'Actual_sales', 'yhat': 'Forecast_Sales', 
                                'yhat_lower': 'Lower_Bound', 'yhat_upper': 'Upper_Bound'}, inplace=True)
```

```python
forecast_output['Forecast_Sales'] = np.expm1(forecast_output['Forecast_Sales'])
forecast_output['Lower_Bound'] = np.expm1(forecast_output['Lower_Bound'])
forecast_output['Upper_Bound'] = np.expm1(forecast_output['Upper_Bound'])
```

```python
forecast_output.drop(columns=['y_log'], inplace=True)
```

```python
forecast_output.head()
```

```python
forecast_output['Error'] = (
    forecast_output['Actual_sales'] - forecast_output['Forecast_Sales']
)
```

```python
forecast_output['APE_%'] = (
    forecast_output['Error'] / forecast_output['Actual_sales']
) * 100
```

```python
# Now Check the Error in model
Predicted = forecast_output.Forecast_Sales
Actual = forecast_output.Actual_sales
```

```python
evaluation_df['Error'] = (evaluation_df['Actual_sales']- evaluation_df['Forecast_Sales'])
evaluation_df['Absolute_Error'] = (evaluation_df['Error'].abs())
evaluation_df['APE_%'] = (evaluation_df['Absolute_Error'] / evaluation_df['Actual_sales']) * 100
```

```python
mae = evaluation_df['Absolute_Error'].mean()
rmse = np.sqrt((evaluation_df['Error'] ** 2).mean())
mape = evaluation_df['APE_%'].mean()
```
